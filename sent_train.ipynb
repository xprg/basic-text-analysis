```python
from keras.datasets import imdb

(xtrain,ytrain),(xtest,ytest)=imdb.load_data(num_words=5000)

len(xtrain)

```

```python
word_index=imdb.get_word_index()
```

```python
index_word=dict([idx,word] for (word,idx) in word_index.items())
```

```python
import numpy as np

def vectorize(sentences,dim=5000):
   output=np.zeros((len(sentences),dim))

   for (i,idx) in enumerate(sentences):
     output[i,idx]=1

   return output

```

```python
xtrain=vectorize(xtrain)
xtest=vectorize(xtest)
```

```python
ytrain=np.asarray(ytrain).astype('float32').reshape(-1,1)
ytest=np.asarray(ytest).astype('float32').reshape(-1,1)
```

```python
from keras import models
from keras.layers import Dense
```

```python
model=models.Sequential()
model.add(Dense(16,activation='relu',input_shape=(5000,)))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

training



```python
x_val=xtrain[:5000]
xtrain_new=xtrain[5000:]

y_val=ytrain[:5000]
ytrain_new=ytrain[5000:]
```

```python
model.summary()
```

```python
hist=model.fit(xtrain_new,ytrain_new,epochs=20,batch_size=256,validation_data=(x_val,y_val))
```

```python
model.evaluate(xtest,ytest)
```
